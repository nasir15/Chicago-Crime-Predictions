```python
#So now we are ready to do Crime Predictions , these predictions will be done by some Machine learning algos
#So we will do some label encoding on these columns because Machine learning algos dont work on Strings
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
#Now we will copy our data frame to a new DaTA Frame
ndf=df.copy()
```

```python
ndf.head()
```

```python
#Label encoding on Date column of new data frame 'ndf'
ndf['Date']=le.fit_transform(ndf['Date'])
```

```python
#Label encoding on IUCR column of new data frame 'ndf'
ndf['IUCR']=le.fit_transform(ndf['IUCR'])
```

```python
#Label encoding on Priamry Type column of new data frame 'ndf'
ndf['Primary Type']=le.fit_transform(ndf['Primary Type'])
```

```python
#Label encoding on Description column of new data frame 'ndf'
ndf['Description']=le.fit_transform(ndf['Description'])
```

```python
#Label encoding on Location Description column of new data frame 'ndf'
ndf['Location Description']=le.fit_transform(ndf['Location Description'])
```

```python
#Label encoding on Arrest column of new data frame 'ndf'
ndf['Arrest']=le.fit_transform(ndf['Arrest'])
```

```python
#Label encoding on Domestic column of new data frame 'ndf'
ndf['Domestic']=le.fit_transform(ndf['Domestic'])
```

```python
#Now we check the new data frame , are all the coulmns of ndf data frame labeled endoded or not ?
```

```python
#New Data frame
ndf.head()
```

```python
#Old Data Frame
df.head()
```

```python
#So we can easily check that our all columns label encoded
#Now we will split our data into a training and test data
```

```python
#Here we import train_test_split to divide our data into train and test 
from sklearn .model_selection import train_test_split
```

```python
#Here we divide our data columns (predictors)
X=ndf[['Date','IUCR','Description','Location Description','Arrest','Domestic']]
```

```python
#This will be the predictions
Y=ndf['Primary Type']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#Here we will apply Rndom Forest Machine learning algorthim
from sklearn.ensemble import RandomForestRegressor
```

```python
#Here we choose n_estimators and random state
model = RandomForestRegressor(n_estimators = 10, random_state = 42)
```

```python
model.fit(X_train,Y_train)
```

```python
#This is the predictions that we predicted from ourtest data
predictions = model.predict(X_test)
```

```python
predictions
```

```python
Y_test
```

```python
predictions2 = [round(x) for x in predictions]
```

```python
#Now we make our new data frame
submissions=pd.DataFrame()
```

```python
#Assign our Test Data columns to a new Data Frame that will predict the Crime Types in Chicago
submissions[['Date','IUCR','Description','Location Description','Arrest','Domestic']]=X_test
```

```python
#Assign all predictions of test Data to the new Data frame
submissions['Predicted Crime Type']=predictions2
```
